```python
line_word = line.replace("\r\n", "")
#line_word = line_word.replace(".", ". ")
line_word = line_word.replace("\"", "")
#line_word = line_word.replace("\"", " \" ")
line_word = line_word.replace("  ", " ")
word_list = line_word.split(" ")
```

```python
word_set = list(set(word_list))
print(len(word_list))
```

```python
word_pos = list()
for word in word_list:
    word_pos.append(word_set.index(word))
print(word_pos)
```

```python
#word_pos_one_hot = tf.one_hot(word_pos, depth=len(word_set))
#word_x = word_pos_one_hot[:len(word_set)-1]
word_x = word_pos[:len(word_set)-1]
print(word_x)
```

```python
word_y = word_pos[1:]
```

```python
learning_rate = 0.001
total_epoch = 30
batch_size = 100
n_step = len(word_set)
n_input = len(word_set[0])
n_class = n_input
n_hidden = 100

x = tf.placeholder(tf.float32, [None, None, n_input])
y = tf.placeholder(tf.int32, [n_class])

x = x.set_shape(None, None, n_input)

sw = tf.get_variable(name="sw", shape=[n_hidden, n_class], initializer=tf.contrib.layers.xavier_initializer())
sb = tf.get_variable(name="sb", shape=[n_class], initializer=tf.contrib.layers.xavier_initializer())
```

```python
cell = tf.nn.rnn_cell.GRUCell(n_hidden)
outputs, _ = tf.nn.dynamic_rnn(cell, x, dtype=tf.float32)
outputs = tf.transpose(outputs, [1,0,2])
outputs = outputs[-1]
model = tf.add(tf.matmul(outputs, sw), sb)
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=model, labels=y))
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```
